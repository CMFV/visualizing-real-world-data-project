```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()

g_key= os.environ["Token_maps"]

lat=41.896516
long=-87.643622
rad=5000
typ='school'

res=requests.get('https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={},{}&radius={}&type={}&key={}'.format(lat,long,rad,typ,g_key)).json()
res
```

```python
schools = []
schools_lat = []
schools_lng = []
resultados=res['results']
for i in range(len(resultados)):
    noun=res["results"][i]["name"]
    lat=res["results"][i]["geometry"]["location"]["lat"]
    lng=res["results"][i]["geometry"]["location"]["lng"]
    schools.append(noun)
    schools_lat.append(lat)
    schools_lng.append(lng)
colegios=pd.DataFrame(schools)
colegios['lat']=schools_lat
colegios['lng']=schools_lng
display(colegios)
```
