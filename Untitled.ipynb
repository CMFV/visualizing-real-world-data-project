```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client.companies_clean
```

```python
import pandas as pd
data = db.collection_companies_clean.find({"category_code": {"$eq": "tech_company"},"founded_year":{"$gt": 2006}})
data0 = pd.DataFrame(data)
data0
```

```python
import json as json
data0.to_json('./geocompanies.json', orient='records', default_handler = str)
```

```python
def listCoord(columna):
    lista_geop=[]
    for i in range(len(data0)):
        lista_geop.append(data0.oficina_principal[i])
    return lista_geop

lista_coordenadas= listCoord(data0.oficina_principal)

lista_coordenadas
```

```python
def officesCount(lista_coordenadas, radio):
    return list(db.collection_companies_clean.find({
        "oficina_principal": {
         "$near": {
           "$geometry":lista_coordenadas,
           "$maxDistance": radio
         }
       }
    }))
radio=2000
N_oficinas_cercanas=[]
for i in range(len(data0)):
    numero_oficinas=officesCount(lista_coordenadas[i], radio)
    N_oficinas_cercanas.append(len(numero_oficinas))

print(N_oficinas_cercanas)
```

```python
data0['near_offices']=N_oficinas_cercanas
display(data0)
```
