```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

```python
import pandas as pd
data = db.companies.find({
    "offices":{
        "$not":{
            "$size":0
        }
    }})

data0 = pd.DataFrame(data)
data0.head()
```

```python

def getFirst(data):
    data = data['offices']
    principal = None
    if data[0]['latitude'] and data[0]['longitude']:
        principal = {
            "type":"Point",
            "coordinates":[data[0]['longitude'], data[0]['latitude']]
        }

    return {
        "totalOffices": len(data),
        "lat": data[0]['latitude'],
        "lng": data[0]['longitude'],
        "oficina_principal": principal
    }


data_office = data0[["offices"]].apply(getFirst, result_type="expand", axis=1)
data_office.head()
```

```python
df_clean = pd.concat([data0,data_office], axis=1)
df_clean.head()
```

```python
df_clean.dropna(subset=['lat'], inplace=True) 
df_clean.head()
```

```python
print(df_clean.columns.values)
```

```python
df_clean["category_code"].value_counts()
```

```python
import re
def business_catergoty(m):
    m = str(m)
    if (re.search('\s*(web)|(software)|(games_video)|(mobile)|(ecommerce)|(network_hosting)|(biotech)|(cleantech)|(analytics)|(nanotech)\s*', m)):
        return 'tech_company'
    else: return m



df_clean["category_code"]=df_clean["category_code"].apply(business_catergoty)

df_clean.head()
```

```python
df_clean1=df_clean[df_clean['deadpooled_year'].isna()]
df_clean1.head()
```

```python
df_clean1["total_money_raised"].value_counts()
```

```python
def money_normalize(mo):
    if 'M' in mo:
        mo=mo.replace('M', '')
        return float(mo)*1000000
    elif 'k' in mo:
        mo=mo.replace('k', '')
        return float(mo)*1000
    elif 'B' in mo:
        mo=mo.replace('B', '')
        return float(mo) *1000000000
    else: return mo
    
def currency_normalize(mo):
    currencies= {
        'CAD': 1.3055728934, 'HKD': 7.8004458315, 'ISK': 126.348640214,'PHP': 51.3419527419, 'DKK': 6.6545697726,
        'HUF': 289.4872938029, 'CZK': 22.7454302274, 'GBP': 0.7991529202, 'RON': 4.2086491306, 'SEK': 9.4492197949,
        'IDR': 14110.0044583148, 'INR': 68.5648684797, 'BRL': 3.8120374498, 'RUB': 63.7561301828, 'HRK': 6.5942933571,
        'JPY': 108.6045474811, 'THB': 30.7748551048,'CHF': 0.992153366,'EUR': 0.8916629514,'MYR': 4.1405260811, 'BGN': 1.7439144004,
        'TRY': 5.7337494427, 'CNY': 6.8812304949, 'NOK': 8.6348640214, 'NZD': 1.5058403923, 'ZAR': 14.1607668301, 'USD': 1.0,
        'MXN': 18.9550601872,'SGD': 1.359518502,'AUD': 1.4318323674,'ILS': 3.5751226037,'KRW': 1178.8408381632,'PLN': 3.7922425323
    }
    if "C$" in mo:
        return money_normalize(mo.replace('C$',''))*currencies["CAD"]
    elif "$" in mo:
        return money_normalize(mo.replace('$',''))
    elif "kr" in mo:
        return money_normalize(mo.replace('kr',''))*currencies["KRW"]
    elif "r" in mo:
        return money_normalize(mo.replace('C',''))*currencies["RUB"]
    elif "€" in mo:
        return money_normalize(mo.replace('€',''))*currencies["EUR"]
    elif "£" in mo:
        return money_normalize(mo.replace('£',''))*currencies["GBP"]
    else:return mo
 


df_clean1["total_money_raised"]=df_clean1["total_money_raised"].apply(currency_normalize)

```
